# Deep Learning - Máquina Restrita de Boltzmann (RBM)

A Máquina Restrita de Boltzmann (RBM - Restricted Boltzmann Machine) é um tipo de modelo de aprendizado de máquina que faz parte da família das redes neurais profundas. Ela é usada principalmente em tarefas de aprendizado não supervisionado, como redução de dimensionalidade, recomendação e geração de dados.

## Base MNIST

A base contém 60 mil imagens com as dimensões 28x28 = 784 pixels

```python
import random
random.seed(0)
```

```python
from tensorflow.keras.datasets import mnist  # Importa a base de dados MNIST do TensorFlow

# Carrega os dados de treinamento e teste da base MNIST
(train_X, train_y), (test_X, test_y) = mnist.load_data()

# Exibe as primeiras 5 imagens da base de treinamento
for i in range(5):
    plt.subplot(1, 5, i + 1)
    plt.imshow(train_X[i], cmap='gray')
    plt.title(f"Label: {train_y[i]}")
    plt.axis('off')

plt.show()
```

## Tratando a Base

A função Bernoulli RBM considera que as colunas das features estejam entre [0,1], a base MNIST está em uma escala entre [0,255].

Precisamos ajustar a escala

```python
def scale(X, eps = 0.001):
    # scale the data points s.t the columns of the feature space
    # (i.e the predictors) are within the range [0, 1]
    return (X - np.min(X, axis = 0)) / (np.max(X, axis = 0) + eps)
```

```python
X_train = scale(train_X)
```

Redimensiona para o formato exigido pelo modelo (duas dimensões) 

1a Observaçoes (imagens) <br>
2a Pixels (28x28 = 784)

```python
def redimensiona(X):
    return X.reshape(-1,784)
```

```python
X_train = redimensiona(X_train)
X_train.shape
```

```python
from sklearn.neural_network import BernoulliRBM
rbm = BernoulliRBM(n_components=100, 
                   learning_rate=0.01, 
                   n_iter = 10, 
                   batch_size = 10,
                   random_state=0, 
                   verbose=True)
rbm.fit(X_train)
```

## Comparando com Base Teste

```python
# Rescala entre 0 e 1 e redimensiona
X_test = scale(test_X)
X_test = redimensiona(X_test)
```

```python
X_test.shape
```

```python
print("Training set Pseudo-Likelihood =", rbm.score_samples(X_train).mean())
print("Validation set Pseudo-Likelihood =", rbm.score_samples(X_test).mean())
```

### Reconstruindo a Base Teste

Referência: https://ml-lectures.org/docs/unsupervised_learning/Denoising.html

```python
# Perform the denoising
k_iter = 12  # Number of Gibbs Sampling Iterations
alpha = 0.9  # Decay factor for the averaging

# Gibb sampling steps
b = rbm.gibbs(X_test)
x_final = np.zeros(784) + np.copy(b) # reconstroi uma matriz zerada do tamanho da base de Teste
for i in range(k_iter):
    b = rbm.gibbs(b)
    x_final += (alpha**(i+1))*b.astype(float) # Averaging the images
```

```python
# Exibe as imagens reais da base de teste e as imagens geradas pelo RBM

num_generated_images = 10

plt.figure(figsize=(10, 5))
for i in range(num_generated_images):
    plt.subplot(2, num_generated_images, i + 1)
    plt.imshow(X_test[i].reshape(28, 28), cmap='gray')
    plt.title("Real")
    plt.axis('off')

    plt.subplot(2, num_generated_images, num_generated_images + i + 1)
    plt.imshow(x_final[i].reshape(28, 28), cmap='gray')
    plt.title("Generated")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

O teste de comparação entre as imagens serve apenas para verificar se o modelo está ajustado. O que interessa é a capacidade de reconstrução dos inputs.

### Pesos 

```python
plt.figure(figsize=(20, 20))
for i, comp in enumerate(rbm.components_):
    plt.subplot(10, 10, i + 1)
    plt.imshow(comp.reshape((28, 28)), cmap='gray',
               interpolation='nearest', vmin=-2.5, vmax=2.5)
    plt.axis('off')
plt.suptitle('100 components extracted by RBM', fontsize=16);
```

### Entendendo a função gibbs

```python
digits_new = np.zeros(784) + np.copy(b)
```

```python
# gibbs sampling here
for j in range(10):
    for i in range(784):
        digits_new[i,:] = rbm.gibbs(digits_new[i,:])
```

```python
num_generated_images = 10

plt.figure(figsize=(10, 5))
for i in range(num_generated_images):
    plt.subplot(2, num_generated_images, i + 1)
    plt.imshow(X_test[i].reshape(28, 28), cmap='gray')
    plt.title("Real Image")
    plt.axis('off')

    plt.subplot(2, num_generated_images, num_generated_images + i + 1)
    plt.imshow(digits_new[i].reshape(28, 28), cmap='gray')
    plt.title("Generated Image")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
