# Construindo Rede Neural

### Etapas

- 1. Arquitetura da rede.
- 2. Inicializa um vetor de historia da função custo para guardar resultados.
- 3. Faz foward loop.
- 4. Calcula perda.
- 5. Atualiza parâmetros.
- 6. Usa os novos parâmetros.

A quantidade de iterações é chamada de épocas (epochs). <br> É a quantidade de vezes que se repetem os passos 1 a 6 para definir os melhores pesos

```python
# Setando o seed da randomização
np.random.seed(0)
```

## Transpondo as Bases

```python
X_train = X_train.transpose()
X_test = X_test.transpose()
y_train = pd.DataFrame(y_train).transpose()
y_test = pd.DataFrame(y_test).transpose()
```

# Construindo a Rede

### Arquitetura da Rede

```python
# Arquitetura da Rede

def getLayerSize (X, y, hidden_neurons):
    n_x = len(X) # quantidade de linhas de x = neurônios da camada de entrada
    n_h = hidden_neurons # quantidade de neurônios na camada escondida
    n_y = len(y) # quantidade de linhas de y = neurônios da camada de saída
    
    size = {'n_x': n_x, 
            'n_h': n_h, 
            'n_y':n_y}
    
    return size
```

```python
layer_size = getLayerSize(X_train, y_train, 4)
```

```python
layer_size
```

### Definindo os Pesos

Inicia Parâmetros

Chute dos Valores dos Pesos Seguindo uma dsitribuição uniforme

```python
import random

def initializeParameters (X, layer_size):
    n_x = layer_size['n_x']
    n_h = layer_size['n_h']
    n_y = layer_size['n_y']
    
    #Chute dos Pesos (quantidade de neurônios na camda de entrada * neuronios camada escondida)
    w1 = [[np.random.randn() for _ in range(n_x)] for _ in range(n_h)] # colunas e linhas
    w2 = [[np.random.randn() for _ in range(n_h)] for _ in range(n_y)] # colunas e linhas
    
    params = {'w1': w1, 
              'w2': w2}
    
    return params
```

```python
init_params = initializeParameters(X_train, layer_size)
```

```python
init_params
```

### Função de Ativação

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

### Forward Propagation

Aplicar as funções de aplicação dos valores que chegam nos neurônios

```python
def forwardPropagation (X, params, layer_size):
    
    n_h = layer_size['n_h']
    n_y = layer_size['n_y']
    
    w1 = params['w1']
    w2 = params['w2']
    
    Z1 = np.dot(w1, X) # multiplica os pesos pelas variáveis
    A1 = sigmoid(Z1) # aplica a função de ativação
    Z2 = np.dot(w2, A1)
    A2 = sigmoid(Z2)
    
    fwd_prop = {'Z1':Z1, 
             'A1': A1, 
             'Z2': Z2, 
             'A2': A2}
    
    return fwd_prop
```

```python
fwd_prop = forwardPropagation(X_train, init_params, layer_size)
```

```python
fwd_prop
```

### Função Custo

Verificar o quanto estamos errando e com base nisso atualizar os pesos

```python
def computeCost (y, fwd_prop):
    
    m = len(y_train.columns)
    
    A2 = fwd_prop['A2']
    
    cost = sum(((y-A2).transpose().sum())**2)/m
    
    return cost
```

```python
computeCost(y_train, fwd_prop)
```

### Back Propagation

Ajustes dos Pesos

```python
def backwardPropagation (X, y, fwd_prop, params, layer_size):
    
    m = len(X.columns)
    
    n_x = layer_size['n_x']
    n_h = layer_size['n_h']
    n_y = layer_size['n_y']
    
    A2 = fwd_prop['A2']
    A1 = fwd_prop['A1']
    w2 = params['w2']
    
    dz2 = A2 - y
    dw2 = 1/m * np.dot(dz2, list(zip(*A1))) # o quanto vamos alterar no peso 2 em função do erro dz2
    
    dz1 = np.dot(list(zip(*w2)), dz2) * (1 - A1**2)
    dw1 = 1/m * np.dot(dz1, X.transpose())
    
    grads = {'dw1': dw1, 
             'dw2': dw2}
    
    return grads
```

```python
grads= backwardPropagation(X_train, y_train, fwd_prop, init_params, layer_size)
```

```python
grads
```

### Alterando os Pesos

Efetivamente alterar os pesos com base no backward Propagation

```python
def updateParameters (grads, params, learning_rate):
    
    w1 = params['w1']
    w2 = params['w2']
    
    dw1 = grads['dw1']
    dw2 = grads['dw2']
    
    w1 = w1 - learning_rate * dw1 # peso anterior menos taxa de aprendizado vezes erro residual
    w2 = w2 - learning_rate * dw2
    
    updated_params = {'w1': w1, 
                      'w2': w2}
    
    return updated_params
```

```python
init_params = updateParameters(grads, init_params, 0.01)
```

```python
init_params
```

### Treina o Modelo

Efetivamente treina o modelo para chegar no melhor peso

```python
def trainModel (X, y, num_iteration, hidden_neurons, lr):
    
    layer_size = getLayerSize(X, y, hidden_neurons) # cria a arquitetura
    init_params = initializeParameters(X, layer_size) # chuta os valores dos pesos
    
    cost_history = []
    
    for i in range(num_iteration):
        fwd_prop = forwardPropagation(X, init_params, layer_size)
        cost = computeCost(y, fwd_prop)
        back_prop = backwardPropagation(X, y, fwd_prop, init_params, layer_size)
        update_params = updateParameters(back_prop, init_params, learning_rate= lr)
        init_params = update_params
        cost_history.append(cost)
        
    model_out = {'updated_params': update_params, 
                 'cost_history': cost_history}
    
    return model_out
```

### Aplica o Treinamento

```python
epocas = 100
hidden_neurons = 40
learning_rate = 0.01

train_model = trainModel(X_train, y_train, epocas, hidden_neurons, learning_rate)
```

## Testa o Modelo

O modelo é bastante ruim, o intuíto é entender o funcionamento da Rede Neural

```python
layer_size = getLayerSize(X_test, y_test, hidden_neurons)
params = train_model['updated_params']
fwd_prop = forwardPropagation(X_test, params, layer_size)
y_pred = pd.DataFrame(fwd_prop['A2']).transpose()

compare = pd.DataFrame()
compare['y_pred'] = y_pred
compare['quality'] = y_test.transpose().reset_index().drop('index', axis=1)
```

```python
compare
```
