## Objetivo

Prever a Humidade às 3pm baseado na Humidade às 9am para a cidade de Albury

```python
df = pd.read_csv(r'bases\weatherAUS.csv')
df
```

```python
df_albury = df[df['Location'] == 'Albury']
df_albury = df_albury[['Humidity9am','Humidity3pm']]
df_albury
```

```python
df_albury.describe()
```

```python
df_albury.info()
```

### Removendo Nulos

```python
df_cleaned = df_albury.dropna()
```

```python
df_cleaned.info()
```

### Tratando a Base para ser usada no Modelo

```python
df_used = df_cleaned.iloc[:3000,:]
df_used
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df_used['X_std'] = scaler.fit_transform(df_used['Humidity9am'].values.reshape(-1,1))
df_used['y_std'] = scaler.fit_transform(df_used['Humidity3pm'].values.reshape(-1,1))
df_used
```

### Transforma os Dados em uma Matriz 30x100

A ideia é enviar várias amostras diversas vezes para manter a recorrência na rede. <br>
Para aprender conforme as amostras vão passando.

A base precisa ter 3 Dimensões

- 1- Quantidade de vezes que irá passar a amostra
- 2- Amostras
- 3- Número de features

```python

X = df_used['X_std'].values.reshape(100,30,1)
y = df_used['y_std'].values.reshape(100,30,1)
```

```python
# Train Test Split Manual
X_train = X[:80,:]
X_test = X[80:,:]

y_train = y[:80,:]
y_test = y[80:,:]
```

```python
len(X_train), len(X_test)
```

```python
len(y_train), len(y_test)
```

## Aplicando Modelo RNN

Documentação: https://www.tensorflow.org/api_docs/python/tf/keras/layers/SimpleRNN

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, LSTM
```

```python
from keras.optimizers import Adam
#adam_lr = Adam(learning_rate=0.001)

# Camada RNN
model = Sequential()
model.add(SimpleRNN(units=15, activation='relu'))

# Camada de saída
model.add(Dense(units=30)) # saída 30 pois temos 30 observações em cada sequência 

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
history = model.fit(X_train, y_train, epochs=100, batch_size=30)
```

```python
history_df = pd.DataFrame(history.history)
history_df.plot();
```

## Prevendo Modelo teste

```python
def indicadores (df, y, y_hat):
    n = len(df)
    SQE = sum((y - y_hat)**2)
    QME = SQE/n
    
    SST = sum((y - y.mean())**2)
    QMT = SST/n
    
    R2 = 1-SQE/SST

    print(f'SQE: {SQE}')
    print(f'QME: {QME}')
    print(f'SST: {SST}')
    print(f'QMT: {QMT}')
    print(f'R2: {R2}')
```

```python
y_hat_test = model.predict(X_test)
```

```python
y_test.shape, y_hat_test.shape
```

```python
result = pd.DataFrame()
result['y_test'] = y_test.reshape(600)
result['y_hat'] = y_hat_test.reshape(600)
result
```

```python
indicadores(result, result['y_test'], result['y_hat'])
```

R2 de 48% e médias próximas

```python
result.describe()
```

```python
plt.style.use('seaborn')
plt.plot('y_test' ,data=result, label = 'Humidade Real')
plt.plot('y_hat' ,data=result, label = 'Previsto')
plt.legend()
```
